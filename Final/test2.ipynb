```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk



```

```python
# Calculate average ratings
# Read the CSV file
data = pd.read_csv('teams_players.csv')  # Replace 'your_file.csv' with the actual filename

# Concatenate home and away ratings for each team
team_ratings = pd.concat([data[['home_team_api_id', 'home_rating']].rename(columns={'home_team_api_id': 'team_id', 'home_rating': 'rating'}),
                          data[['away_team_api_id', 'away_rating']].rename(columns={'away_team_api_id': 'team_id', 'away_rating': 'rating'})])

# Calculate average ratings for each team
avg_ratings = team_ratings.groupby('team_id')['rating'].mean()

# histogram of average ratings
avg_ratings.hist(bins=15)
plt.xlabel('Average Rating')
plt.ylabel('Number of Teams')
plt.title('Distribution of Average Ratings')
plt.show()
```
